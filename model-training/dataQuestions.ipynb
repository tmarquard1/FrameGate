```python
import tensorflow as tf
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the CSV file
csv_file_path = '../data-generation/impact.csv'
data = pd.read_csv(csv_file_path)
```

```python
# Extract the percieved change and the top 1 predictions for frame n and m
# Extract the relevant columns
extracted_data = data[['Perceived Change', 'Top 1 Label (n)', 'Top 1 Probability (n)', 'Top 1 Label (m)', 'Top 1 Probability (m)']]

# Display the extracted data
print(extracted_data.head())
```

```python
# Plot the normalized histogram of the percieved change
# Plot the normalized histogram of the perceived change
plt.hist(extracted_data['Perceived Change'], bins=100, density=True)
plt.xlabel('Perceived Change')
plt.ylabel('Density')
plt.title('Normalized Histogram of Perceived Change')
plt.show()
```
